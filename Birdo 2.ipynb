```python
#create a high resolution working space

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300
mpl.rcParams['font.family']= 'serif'

#import modules
import helpers as h
import pandas as pd
import matplotlib.pyplot as plt

#read data file from stoashi
df = pd.read_csv('data.csv')
#print(df)

L = 135 #mm from spec

V, D1, D2 = df['V'], df['D1'], df['D2']

Voltage1, sinAngle1 = h.getValues(D1, L, V)
#print(Voltage1,sinAngle1)
Voltage2, sinAngle2 = h.getValues(D2, L, V)
#print(Voltage2,sinAngle2)

#check work
plt.figure(1)
plt.plot(Voltage1,sinAngle1,label='D1')
plt.plot(Voltage2,sinAngle2,label='D2')
plt.legend()

#print slope
slope1 = h.linFit(Voltage1,sinAngle1)
slope2 = h.linFit(Voltage2,sinAngle2)
print(slope1,slope2)

#find theoretical values for D1 and D2 based on slopes
D1_theor = h.getD(slope1)
D2_theor = h.getD(slope2)
print(D1_theor,D2_theor)
print(D1_theor/D2_theor)
```
